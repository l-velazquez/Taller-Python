# Numpy
## Max, Min, Mean, Median, Standard Deviation

```python

print(np.max(age))
print(np.min(age))
print(np.mean(age))
print(np.median(age))
print(np.std(age))

print("\n")
print(age.describe())
```

# Histograms

```python
plt.hist(age,bins= 80)
plt.title("Age")
```

```python
plt.hist(g,bins=25)
```

# Scatter Plot

```python

plt.figure(figsize=(16,10))
plt.scatter(age,glucosa)
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.title("Age vs Glucose Value")
plt.show()
```

```python
df2 = df[['age','avg_glucose_level']]
```

# KMeans Clustering - Machine Learning

```python

kmeans = KMeans(2)
kmeans.fit(df2)
id = kmeans.predict(df2)
```

```python

plt.figure(figsize=(14,8))
plt.scatter(df2['age'],df2['avg_glucose_level'],c=id)
```
