## Usando kmeans para clusterisar la data (General)

```python
from  sklearn.metrics import pairwise_distances_argmin_min
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# load the dataset
X = df.copy()

# select the columns to use for clustering
X = df[['age','gender_M','1975-2_Labt-test A']]

# create a KMeans model with 2 clusters
kmeans = KMeans(n_clusters=2)
# create a StandardScaler object
scaler = StandardScaler()

# fit the scaler to the data and transform the data
X_scaled = scaler.fit_transform(X)

# fit the model to the data
kmeans.fit(X_scaled)

# predict the cluster labels for the data
labels = kmeans.predict(X_scaled)

# get the centroids of the clusters
centroids = kmeans.cluster_centers_

# reduce the data to 2 dimensions using PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)
# create a scatter plot of the reduced data, colored by cluster label
plt.title('PCA_2 Compontent: Cluster 2')
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=labels)

plt.show()
plt.title('Gender X 1975-2_Labt-test A')
plt.xlabel('Gender')
plt.ylabel('1975-2_Labt-test A')
plt.scatter(X['gender_M'],X['1975-2_Labt-test A'],c=labels)
```

## Centros del cluster ( usualmente representan la diferencia mas grande encontrada) (General)

```python
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, X)
X['clusters'] = labels
X.iloc[closest]
```

## Describiendo la data divida por clusters (General)

```python
Cluster_0_group = X.groupby('clusters').get_group(0)
Cluster_0_group.describe().loc[:,['gender_M','1975-2_Labt-test A']]
```

```python
Cluster_1_group = X.groupby('clusters').get_group(1)
Cluster_1_group.describe().loc[:,['gender_M','1975-2_Labt-test A']]
```

# K-Means Clustering K = 3:

```python

def kmeans_group_calculator(group_range,clusters,limits):

    # create variable to hold centroids
    centroids = pd.DataFrame()

    # create n figures (the amount of groups needed) to insert each group value
    fig, axs = plt.subplots(1, group_range[1]-group_range[0], figsize=(25,10))
    plt.subplots_adjust(wspace=0.35) # set the width space between subplots to be 0.35

    for group_id,ax in zip(range(*group_range),axs):

        # load the dataset
        group_x_data = df[df["groups"] == group_id]

        # select the columns to use for clustering
        group_x_data = group_x_data[['1975-2_Labt-test A']]

        # create a KMeans model with 3 clusters
        kmeans = KMeans(n_clusters=clusters)

        # create a StandardScaler object
        scaler = StandardScaler()

        # fit the scaler to the data and transform the data
        X_scaled = scaler.fit_transform(group_x_data)

        # fit the model to the data
        kmeans.fit(X_scaled)

        # predict the cluster labels for the data
        labels = kmeans.predict(X_scaled)

        # get the centroids of the clusters
        closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, group_x_data)
        centroid = group_x_data.iloc[closest]
        centroid = centroid.reset_index(drop=True)
        centroid = centroid.rename(columns={'1975-2_Labt-test A': '1975-2_Labt-test A'+str(group_id)})
        centroids['1975-2_Labt-test A'+str(group_id)] = centroid

        # display range of values
        group_x_data['clusters'] = kmeans.labels_
        
        for name,group in group_x_data.groupby('clusters'):
            print(f'Group: {group_id} | Cluster {name}')
            print(f'1975-2_Labt-test A range: {group["1975-2_Labt-test A"].min()} - {group["1975-2_Labt-test A"].max()}')
        print('')
        # set y limits to be equal between subplots
        ax.set_ylim(*limits)

        # generate scatterplot
        sns.scatterplot(x=group_id,y=group_x_data['1975-2_Labt-test A'],hue=labels, ax=ax)
    

        # generate legend
        ax.legend()

        # set both x label and y label to blank (will be set globally later)
        ax.set_xlabel('')
        ax.set_ylabel('')

    # global tittle
    plt.suptitle('Group ID X 1975-2_Labt-test A')

    # global x and y labels
    fig.text(0.1, 0.5, '1975-2_Labt-test A', ha='center', va='center', rotation='vertical')
    fig.text(0.5, 0, 'Group ID', ha='center', va='center')

    # return centroid from each group
    return centroids

# recieve centroids
centroids = kmeans_group_calculator((1,9),3,(0.0,2.0))
```

```python
from sklearn.metrics import silhouette_score
def dbscan_group_calculator(group_range,epsmin,limits):

    # create variable to hold centroids
    centroids = pd.DataFrame()

    # create n figures (the amount of groups needed) to insert each group value
    fig, axs = plt.subplots(1, group_range[1]-group_range[0], figsize=(25,10))
    plt.subplots_adjust(wspace=0.35) # set the width space between subplots to be 0.35

    for group_id,ax in zip(range(*group_range),axs):

        # load the dataset
        group_x_data = df[df["groups"] == group_id]

        # select the columns to use for clustering
        group_x_data = group_x_data[['1975-2_Labt-test A']]

        # create a StandardScaler object
        scaler = StandardScaler()

        # fit the scaler to the data and transform the data
        X_scaled = scaler.fit_transform(group_x_data)

        # create a DBSCAN model
        dbscan = DBSCAN(eps=epsmin[0],min_samples=epsmin[1])

        # fit the model to the data
        dbscan.fit(X_scaled)

        # predict the cluster labels for the data
        labels = dbscan.labels_

        # display range of values
        group_x_data['clusters'] = labels

        # display 
        score = silhouette_score(group_x_data, labels)
        
        print(f'Silhouette Score: {score}')

        for name,group in group_x_data.groupby('clusters'):
            print(f'Group: {group_id} | Cluster {name}')
            print(f'1975-2_Labt-test A range: {group["1975-2_Labt-test A"].min()} - {group["1975-2_Labt-test A"].max()}\n')

        # set y limits to be equal between subplots
        ax.set_ylim(*limits)

        # generate scatterplot
        sns.scatterplot(x=group_id,y=group_x_data['1975-2_Labt-test A'],hue=labels, ax=ax)
    

        # generate legend
        ax.legend()

        # set both x label and y label to blank (will be set globally later)
        ax.set_xlabel('')
        ax.set_ylabel('')

    # global tittle
    plt.suptitle('Group ID X 1975-2_Labt-test A')

    # global x and y labels
    fig.text(0.1, 0.5, '1975-2_Labt-test A', ha='center', va='center', rotation='vertical')
    fig.text(0.5, 0, 'Group ID', ha='center', va='center')

    # return centroid from each group
    return centroids

# recieve centroids
centroids = dbscan_group_calculator((1,7),(1,15),(0.0,2.0))
```

# Looking for Outliers using Gaussian mixture model
Gaussian mixture model to identify possible outliers in the data. You can consider a sample to be an outlier if its probability of belonging to any of the clusters is below a certain threshold. You can set the threshold based on the specific requirements of your application and the characteristics of your data.


### Data used to analyze:
The use of the gender, lab test result and age where use to conclude the outlier


### Observations:
The results of gaussian change sometime when you run it many times

```python
from sklearn.mixture import GaussianMixture
X = df.copy()
# Compute the probabilities of each sample to belong to each cluster
gmm = GaussianMixture(n_components=4)
#X = X[['1975-2_Labt-test A','age']]
X = X[['gender_M','1975-2_Labt-test A']]
# Fit the model to the data
gmm.fit(X)

probabilities = gmm.predict_proba(X)

# Set a threshold for the minimum probability to consider a sample as an inlier
threshold = 0.8
# Find the samples with a probability below the threshold
outliers = X[np.amax(probabilities, axis=1) < threshold]

# Print the outliers
print(outliers)

# Create a scatter plot
plt.scatter(X['gender_M'], X['1975-2_Labt-test A'], c='b', label='Inliers')
plt.scatter(outliers['gender_M'], outliers['1975-2_Labt-test A'], c='r', label='Outliers')

# Add a legend to the plot
plt.legend()

# Add a title to the plot
plt.title('Gaussian Mixture Clustering')

# Display the plot
plt.show()
```

# GaussianMixture 5-95 intervals and cluster centers (no gender):

```python
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial.distance import cdist


# avg container
general_avg = []
# standar deviation container
general_sd = []

# Std containers
general5 = []
general95 = []
male5 = []
male95 = []
female5 = []
female95 = []


def gaussian_group_calculator_creatinine(group_range,components,limits,df):

    # create variable to hold centroids
    centroids = pd.DataFrame()

    # Percentiles container 
    non_scaled_percentiles = []
    scaled_percentiles = []
    non_scaled_array = []
    general_avg_f = []

    # create n figures (the amount of groups needed) to insert each group value
    fig, axs = plt.subplots(1, group_range[1]-group_range[0], figsize=(25,10))
    plt.subplots_adjust(wspace=0.35) # set the width space between subplots to be 0.35

    for group_id,ax in zip(range(*group_range),axs):

        # load the dataset
        group_x_data = df[df["age-group"] == group_id]

        # select the columns to use for clustering
        group_x_data = group_x_data[['value']] 

        # Create a GMM object with the desired number of components
        gmm = GaussianMixture(n_components=components)

        # create a StandardScaler object
        scaler = MinMaxScaler(feature_range=(0, 1))

        # fit the scaler to the data and transform the data
        X_scaled = scaler.fit_transform(group_x_data)

        print("Number of elements in bin:",group_x_data.shape[0])

        # Fit the GMM to the data
        gmm.fit(X_scaled)

        # predict the cluster assignment for each data point
        labels = gmm.predict(X_scaled)

        # get means of the clusters
        means = gmm.means_

        if group_x_data[labels == 0]['value'].mean() > group_x_data[labels == 1]['value'].mean():
            normal_cluster = group_x_data[labels == 1]
        else:
            normal_cluster = group_x_data[labels == 0]
        
        # Take 30000 samples and get 5 and 95 percentiles of the bins
        for i in range(0,100):

            data = normal_cluster.sample(30000,replace=False,random_state=i)

            # Calculate 5th and 95th percentiles
            p5 = np.percentile(data['value'], 5)
            p95 = np.percentile(data['value'], 95)

            # Append percentiles
            general5.append(p5) 
            general95.append(p95)

        general_avg_f.append(normal_cluster.mean())
        print(f'GROUP {group_id}:\n')
        for i,mean in zip(range(0,components),means):
            
            # calculate euclidean distance between each data point and the cluster means
            distances = cdist(X_scaled, mean.reshape(1,-1))
           
            # find the nearest point to the center
            min_distance_index = np.argmin(distances)
           
            print(f'Cluster Center scaled (euclidean) {i}, Value : {X_scaled[min_distance_index]}')
            print(f'Cluster Center non-scaled (euclidean) {i}, Value : {scaler.inverse_transform(X_scaled[min_distance_index].reshape(1, -1))}\n')
           
            #select the data points that belong to the cluster and get the percentiles
            cluster_data = X_scaled[labels == i]
            non_scaled_cluster_data = group_x_data[labels == i]
            scaled_percentiles.append(f' GROUP {group_id}, cluster {i} percentiles: {np.percentile(cluster_data, [5, 95])} ')
            non_scaled_percentiles.append(np.percentile(non_scaled_cluster_data, [5, 95]))
            non_scaled_array.append(scaler.inverse_transform(X_scaled[min_distance_index].reshape(1, -1))[0])
            
            # get mean
            print("Mean:",non_scaled_cluster_data.mean())
            # get standar deviation
            print("Standard Deviation:",non_scaled_cluster_data.std())

            # calculate the median of each feature for the data points in the first cluster
            medians = np.median(cluster_data, axis=0)
            print(f'Cluster Center scaled (median) {i}, Value : {medians}')
            print(f'Cluster Center non-scaled (median) {i}, Value : {scaler.inverse_transform(medians.reshape(1, -1))}\n')
            


        # set y limits to be equal between subplots
        ax.set_ylim(*limits)

        # generate scatterplot
        sns.scatterplot(x=group_id,y=X_scaled[:,0],hue=labels, ax=ax)
    

        # generate legend
        ax.legend()

        # set both x label and y label to blank (will be set globally later)
        ax.set_xlabel('')
        ax.set_ylabel('')

    # global tittle
    plt.suptitle('Group ID X Creatininte_test')

    # global x and y labels
    fig.text(0.1, 0.5, ' Creatininte_test value', ha='center', va='center', rotation='vertical')
    fig.text(0.5, 0, 'Group ID', ha='center', va='center')

    # return centroid from each group
    return non_scaled_percentiles,scaled_percentiles, non_scaled_array

# recieve centroids
non_scaled_percentiles,scaled_percentiles,non_scaled_array = gaussian_group_calculator_creatinine((1,6),2,(0.0,1.0))

```

```python
from scipy.stats import multivariate_normal
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and covariance of the multivariate normal distribution
mean = np.array([0, 0])
cov = np.array([[1, 0.5], [0.5, 1]])

# Generate x and y values to plot the PDF
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y

# Evaluate the multivariate normal PDF at each point in the grid
rv = multivariate_normal(mean=mean, cov=cov)
z = rv.pdf(pos)

# Plot the PDF of the multivariate normal distribution
fig, ax = plt.subplots(dpi=200)
ax.contour(x, y, z)
ax.set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the mean and covariance matrix of the distribution
mean = np.array([0, 0])
cov = np.array([[1, 0.5], [0.5, 1]])

# Define the range of x and y values for the plot
x, y = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
pos = np.dstack((x, y))

# Compute the joint density of the distribution
inv_cov = np.linalg.inv(cov)
det_cov = np.linalg.det(cov)
diff = pos - mean
exponent = np.sum(-0.5 * np.dot(diff, inv_cov) * diff, axis=2)
joint_density = np.exp(exponent) / (2 * np.pi * np.sqrt(det_cov))

# Plot the joint density as a contour plot
fig, ax = plt.subplots(dpi=200)

ax.contourf(x, y, joint_density, levels=1000)
ax.set_aspect('equal')
plt.show()
```
